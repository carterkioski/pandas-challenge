```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
file_to_load = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(file_to_load)
```

## Player Count

* Display the total number of players


```python
purchase_data = purchase_data.dropna(how='any')
player_data = purchase_data.loc[:,['SN','Age','Gender']]
player_data = player_data.drop_duplicates(subset=['SN'])
player_data["SN"].value_counts().count()
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
df = pd.DataFrame([{'Number of Unique Items': purchase_data['Item Name'].value_counts().count(),
                   'Average Price':purchase_data['Price'].mean(),
                   'Number of Purchases': purchase_data["Purchase ID"].count(),
                   'Total Revenue':purchase_data["Price"].sum()}])
#formatting
df[['Average Price','Total Revenue']] = df[['Average Price','Total Revenue']].applymap('${:,.2f}'.format)
df.head()
```

## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
player_data.groupby(['Gender']).count()
```


## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender




* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
summary_df = pd.DataFrame()
summary_df['Purchase Count']= purchase_data.groupby(['Gender'])['Purchase ID'].count()
summary_df['Average Purchase Price']= purchase_data.groupby(['Gender'])['Price'].mean()
summary_df['Total Purchase Value']= purchase_data.groupby(['Gender'])['Price'].sum()
summary_df['Avg Total Purchase per Person']= purchase_data.groupby(['Gender'])['Price'].sum()/(player_data['SN'].value_counts().count())
#formatting
summary_df[['Average Purchase Price','Total Purchase Value','Avg Total Purchase per Person']] = summary_df[['Average Purchase Price','Total Purchase Value','Avg Total Purchase per Person']].applymap('${:,.2f}'.format)
summary_df.head()
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
#set bins,
bins = [0,9,14,19,24,29,34,39,1000]
bin_labels = ['<10','10-14','15-19','20-24','25-29','30-34','35-39','40+']
player_data['Total Count'] = pd.cut(player_data['Age'],bins,labels=bin_labels)
player_data_group = player_data.groupby(['Total Count'])
#resetting the summary temp dataframe
summary_df = pd.DataFrame()
summary_df['Total Count'] = player_data_group['SN'].count()
summary_df['Percentage of Players'] = summary_df["Total Count"]/player_data["Total Count"].count()*100
summary_df['Percentage of Players'] = summary_df['Percentage of Players'].map('{:,.2f}%'.format)
summary_df
```

## Purchasing Analysis (Age)

* Bin the purchase_data data frame by age


* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
purchase_data['Purchase Count'] = pd.cut(purchase_data['Age'],bins,labels=bin_labels) 

purchase_data_group = purchase_data.groupby(['Purchase Count'])
summary_df = pd.DataFrame()
summary_df['Purchase Count'] = purchase_data_group['SN'].count()
summary_df['Average Purchase Price'] = purchase_data_group['Price'].mean()
summary_df['Total Purchase Value'] = purchase_data_group['Price'].sum()
summary_df['Avg Total Purchase per Person'] = purchase_data_group['Price'].mean()
summary_df[['Average Purchase Price','Total Purchase Value','Avg Total Purchase per Person']] = summary_df[['Average Purchase Price','Total Purchase Value','Avg Total Purchase per Person']].applymap('${:,.2f}'.format)
summary_df
```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
spenders_df = pd.DataFrame()
spenders_df['Purchase Count'] = purchase_data.groupby(purchase_data['SN'])['SN'].count()
spenders_df['Average Purchase Price'] = round(purchase_data.groupby(purchase_data['SN'])['Price'].mean(),2)
spenders_df['Total Purchase Value'] = purchase_data.groupby(purchase_data['SN'])['Price'].sum()
spenders_df=spenders_df.sort_values('Total Purchase Value',ascending=False)
spenders_df[['Average Purchase Price','Total Purchase Value']] = spenders_df[['Average Purchase Price','Total Purchase Value']].applymap('${:,.2f}'.format)
spenders_df.head(5)
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, average item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
item_group = pd.DataFrame()
item_data = purchase_data.loc[:,['Purchase ID','Item ID','Item Name','Price']]
item_data = item_data.groupby(['Item ID','Item Name'])
item_group['Purchase Count'] = item_data['Item ID'].count()
item_group['Price'] = item_data['Price'].mean()
item_group['Total Purchase Value'] = item_data['Price'].sum()
item_group[['Price','Total Purchase Value']] = item_group[['Price','Total Purchase Value']].applymap('${:,.2f}'.format)

item_group = item_group.sort_values('Purchase Count',ascending=False)
item_group.head(5)
```
